```python
from wordcloud import WordCloud, STOPWORDS , ImageColorGenerator
import pandas as pd
import matplotlib.pylab as plt
from PIL import Image
import numpy as np
```

```python
stopwords = set(STOPWORDS)
mask = np.array(Image.open('output/images/air.png'))
```

```python
#we will read the text
data_file = pd.read_csv('data/BA_reviews_cleaned.csv')
#wordcloud
wordcloud = WordCloud(stopwords = stopwords , width=1280, height=720,mask=mask,background_color='Black',contour_color='#CF0012',contour_width=2).generate(''.join(data_file['reviews']))
plt.figure(figsize=(20,10),facecolor='k')
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.tight_layout (pad=0)
```

```python
#saving the image of wordcloud
wordcloud.to_file ('output/wordclouds/reviews_wordcloud_Dark.png')
plt.show()
```
